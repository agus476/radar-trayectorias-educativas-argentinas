### Definicion de funciones e Insercion de datos

**Defenicion de funciones:**

- **load_datasets**: obtenemos el dataframe para distintos datasets, y retornamos el df para cada una de ellas
- **crear_hash_sha256**: Craemos el hash para posibles cruces entre tablas y ademas para tener un id unico para cada registro
- **insert_hash_dataframe**: llamamos a la funcion **crear_hash_sha256** que nos retorna el id_hash y lo insertamos al dataframe
- **insercion_datos**: En esta funcion vamos a insertar los datos de los distintos DF que vamos a ir obteniendo

```python
from google.colab import files
import pandas as pd
```

```python
def load_datasets(nombre_archivo):
  #obtenemos los datos del dataset
  uploaded = files.upload()
  df = pd.read_excel(nombre_archivo)

  return df
```

```python
import hashlib
def crear_hash_sha256(fila):
    texto = f"{fila['anio']}-{fila['provincia']}-{fila['departamento']}-{fila['sector']}-{fila['ambito']}"
    return hashlib.sha256(texto.encode()).hexdigest()
```

```python
def insert_hash_dataframe(df):
  try:
  # recreamos el hash nuevamente
      df.insert(0, 'id_hash', df.apply(crear_hash_sha256, axis=1))
  except Exception as e:
    print('posiblemente ya este creado el hash', e)
```

```python
def insercion_datos(nombre_tabla, df):
  # cargamos el dataframe a la base de datos
  df.to_sql(nombre_tabla, engine, if_exists='replace', index=False)
  return nombre_tabla
```

## Dataset trayectoria por sexo

```python
df_trayectoria_por_sexo = load_datasets('Trayectoria_Final_QA.xlsx')
```

```python
# llamamos a la funcion craer hash y le pasamos el dataframe
insert_hash_dataframe(df_trayectoria_por_sexo)
```

```python
#pasamos nombre de la tabla y el dataframe a ingestar
trayectoria_por_sexo = insercion_datos('trayectoria_por_sexo',df_trayectoria_por_sexo)
```
